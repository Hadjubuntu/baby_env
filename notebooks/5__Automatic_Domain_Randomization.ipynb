# Automatic Domain Randomization (ADR)
Some tests to param ADR solution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define min-Max of a value 
param_var = [0.0, 3.0]
```

```python
total_timesteps = 100
```

```python
results = []
progress_list = []

for t in range(total_timesteps):
    progress = t/total_timesteps
    
    gamma = np.random.normal(loc=progress,scale=0.2)
    value = gamma*(param_var[1] - param_var[0]) + param_var[0]
    value = np.clip(value, param_var[0], param_var[1])
    
    results.append(value)
    progress_list.append(progress)
    
```

```python
plt.plot(np.array(progress_list), results)
plt.xlabel('Progression (t/total_timesteps)')
plt.ylabel('Variable value')
plt.title('Example of complexity growing on variable')
plt.show()
```
